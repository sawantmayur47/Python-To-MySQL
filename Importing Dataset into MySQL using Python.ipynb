```python
# Importing the required libraries
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Connecting to the database
engine = create_engine("mysql+pymysql://root:9969043051@localhost/portfolioproject")
```

```python
# Reading the excel file and creating its dataframe
covid_deaths = pd.read_excel("CovidDeaths.xlsx")
covid_deaths
```

```python
# Reading the excel file and creating its dataframe
covid_vaccinations = pd.read_excel("CovidVaccinations.xlsx")
covid_vaccinations
```

```python
# Interacting with SQL databases through SQLAlchemy library
covid_deaths.to_sql(con = engine, name = 'covid_deaths', if_exists = 'replace', index = False)
```

```python
# Checking if the data is stored in SQL database
engine.execute("SELECT * FROM covid_deaths").fetchall()
```

```python
# Checking if the data is stored in SQL database by reading back the data directly as data frame
pd.read_sql_table("covid_deaths", engine)
```

```python
# Interacting with SQL databases through SQLAlchemy library
covid_vaccinations.to_sql(con = engine, name = 'covid_vaccinations', if_exists = 'replace', index = False)
```

```python
# Checking if the data is stored in SQL database
engine.execute("SELECT * FROM covid_vaccinations").fetchall()
```

```python
# Checking if the data is stored in SQL database by reading back the data directly as data frame
pd.read_sql_table("covid_vaccinations", engine)
```
